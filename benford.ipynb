```r
benlaw<-function(d){log10(1+(1/d))}
benlaw(1)
benlaw(2)
benlaw(3)
benlaw(4)
benlaw(5)
benlaw(6)
benlaw(7)
benlaw(8)
benlaw(9)
sum(benlaw(1:9))
```

```r
plot(1:9,(benlaw(1:9)))
```

```r
install.packages("benford.analysis")
library(benford.analysis)
data<-data.frame(Digit = 1:9, prob =benlaw(1:9))
trends = benford(data$Digit,
                 number.of.digits = 1,
                 discrete = T,
                 sign = "positive")
```

```r
trends
```

```r
plot(trends)
```
